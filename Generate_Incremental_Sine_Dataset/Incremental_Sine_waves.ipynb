```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.arange(0, 0.1, 0.0001);
```

```python
list_x = []
list_y = []
dataset = {}
a = 10
delta = np.random.random()
length_trajectory = time.shape[0]
data = {}
total_samples = 50
frequency  = (np.random.random()*60)*np.ones([total_samples,1])
amplitude  = (np.random.random()*1)*np.ones([total_samples,1])
phase      = (np.random.random()*90)*np.ones([total_samples,1])
for i in range(500):
    delta = np.random.random()
    frequency = frequency+ delta
    amplitude = amplitude+ delta
    y  = np.zeros([total_samples, length_trajectory])
    for j in range(50):
        y[j,:] = amplitude[0]*np.sin(2*np.pi*frequency[0]*time+phase[0])
    print(y.shape)
    data['task'+str(i)] = (y, time, phase, amplitude, frequency)
```
